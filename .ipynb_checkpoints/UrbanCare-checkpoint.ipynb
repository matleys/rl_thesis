# Markov Decision Process Environment

## Agent

```python
class Agent:
    def __init__(self, agent_id, name='Test Agent', age=50, qualifications=[], location=(41.390205, 2.154007), 
                 region='A', gender='F', languages=['spanish'], race='white', availabilities=list(range(18,38))):
        self.agent_id = agent_id
        self.name = name
        self.age = age
        self.qualifications = qualifications # nurse, cpr, massage, physio and/or revalidation
        self.location = location # (lat, lon)
        self.region = region # A, B, C...
        self.gender = gender # male or female (M/F)
        self.languages = languages # spanish, catalan, french and/or english
        self.race = race # white, latin-american, black or asian
        self.availabilities = availabilities # see time slot explanation below
        self.current_availabilities = self.availabilities.copy() # a copy of availabilities that can change when an agent gets assigned to a service
        self.active = True # agents can randomly (de)activate at given times
        self.normal_hours = 8
        self.worked_hours = 0
        self.overtime = 0
        self.idle = 0
    
    def __eq__(self, other):
        if isinstance(other, Agent):
            return self.agent_id == other.agent_id
        return NotImplemented
    
    def work(self, num_hours):
        self.worked_hours += num_hours
        if num_hours > self.normal_hours: # overtime
            self.overtime += (num_hours - self.normal_hours)
        else:
            self.idle += (self.normal_hours - num_hours)
    
    def assign(self, time_slot):
        if self.is_available(time_slot) == False:
            raise Exception
        for t in time_slot:
            self.current_availabilities.remove(t)
    
    def __hash__(self):
        return self.agent_id
    
    def is_available(self, time_slot):
        return all(slot in self.current_availabilities for slot in time_slot)
    
    def reset_availabilities(self):
        self.current_availabilities = self.availabilities.copy()
    
    def reset(self):
        self.worked_hours = 0
        self.overtime = 0
        self.idle = 0
    
    def to_string(self):
        return "Agent " + self.name
```

An agent represents a caregiver as a set of defining features.

<b>ASSUMPTION:</b> a day is represented as 48 intervals 30 mins. A time slot is then defined as a list of 30 min intervals. So slot 1 refers to the interval [00:00 - 00:30], slot 2 refers to [00:30 - 01:00], etc.

Thus for example the list [15, 16, 17, ..., 30] is an abstracted way to represent that an agent is available from slot 15 ([07:00 - 07:30]) until slot 30 ([14:30 - 15:00]). So the agent is available from 07:00 until 15:00. 

A complete list of representations can be found below:

```python
def load_agents(filename):
    agents_df = pd.read_csv(filename, sep=";")
    agents_df = agents_df.where(pd.notnull(agents_df), None)
    
    agents=[]
    for idx in agents_df.index:
        agent_row = agents_df.iloc[idx]

        agent_id = int(agent_row['Id'])
        name = agent_row['Name']
        age = int(agent_row['Age'])

        if agent_row['Qualifications'] is None:
            qualifications = []
        else:
            qualifications = agent_row['Qualifications'].split(";")

        location = (float(agent_row['Location'].split(";")[0].replace(',','.')), float(agent_row['Location'].split(";")[1].replace(',','.')))
        region = agent_row['Region']
        gender = agent_row['Gender']
        languages = agent_row['Languages'].split(";")
        race = agent_row['Race']

        availabilities = []
        availabilities = list(range(int(agent_row['Availabilities'].split(";")[0]), int(agent_row['Availabilities'].split(";")[1])))
        
        agent = Agent(agent_id = agent_id, name = name, age = age, qualifications = qualifications, 
                      location = location, region = region, gender = gender, languages = languages, 
                      race = race, availabilities = availabilities)
        agents.append(agent)
        
    return agents
```

```python
all_agents = load_agents('agents.csv')
```

```python
def reset_agents():
    for agent in all_agents:
        agent.reset()
```

## Service

```python
class Service: 
    def __init__(self, service_id, name='Test Service', age=70, service_type=[], location=(41.390205, 2.154007), 
                 region='A', gender='F', languages=['spanish'], race='white', time_slot=list(range(25,29))):
        self.service_id = service_id
        self.name = name
        self.age = age
        self.service_type = service_type # nurse, cpr, massage, physio and/or revalidation
        self.location = location # (lat, lon)
        self.region = region # A, B, C...
        self.gender = gender # male or female
        self.languages = languages # spanish, catalan, french and/or english
        self.race = race # white, latin-american, black or asian
        self.time_slot = time_slot # time slot in which service is needed
        self.active = True # services can randomly (de)activate at given times
    
    def __eq__(self, other):
        if isinstance(other, Service):
            return self.service_id == other.service_id
        return NotImplemented
    
    def __hash__(self):
        return self.service_id
    
    def to_string(self):
        return "Service " + self.name
```

A service represents an elderly as a set of defining features.

```python
def load_services(filename):
    services_df = pd.read_csv(filename, sep=";")
    services_df = services_df.where(pd.notnull(services_df), None)

    services=[]
    for idx in services_df.index:
        service_row = services_df.iloc[idx]

        service_id = int(service_row['Id'])
        name = service_row['Name']
        age = int(service_row['Age'])
        
        if service_row['Type'] is None:
            service_type = []
        else:
            service_type = service_row['Type'].split(";")

        location = (float(service_row['Location'].split(";")[0].replace(',','.')), float(service_row['Location'].split(";")[1].replace(',','.')))
        region = service_row['Region']
        gender = service_row['Gender']
        languages = service_row['Languages'].split(";")
        race = service_row['Race']

        time_slot = []
        time_slot = list(range(int(service_row['TimeSlot'].split(";")[0]), int(service_row['TimeSlot'].split(";")[1])))
        
        service = Service(service_id = service_id, name = name, age = age, service_type = service_type, 
                      location = location, region = region, gender = gender, languages = languages, 
                      race = race, time_slot = time_slot)
        services.append(service)
        
    return services
```

## Pair

```python
def get_quantiles():
    distances=list()
    for agent in all_agents:
        for service in all_services:
            p = Pair(agent, service)
            distances.append(p.dist)
    distances = np.array(distances)
    return np.quantile(distances, 0.25), np.quantile(distances, 0.5), np.quantile(distances, 0.75)
```

```python
class Pair:
    def __init__(self, agent, service):
        self.agent = agent
        self.service = service
        self.pair_id = self.__key()
        # time_periods are the 'overlap' time interval(s) in which both the agent is available and the service is need
        self.time_periods = [slot for slot in agent.current_availabilities if slot in service.time_slot]
    
    def __key(self):
        return hash((self.service.service_id, self.agent.agent_id))
    
    def __eq__(self, other):
        if isinstance(other, Pair):
            return (self.service.service_id == other.service.service_id) and (self.agent.agent_id == other.agent.agent_id)
        return NotImplemented
    
    def __hash__(self):
        return self.pair_id
    
    def isLegal(self):
        # 3 conditions: times must match, agent must have correct qualifications and both agent and service need to be active
        if self.agent.active == False or self.service.active == False:
            return False
        
        if(len(self.time_periods) != len(self.service.time_slot)):
            return False # if there are no time periods, the pair is not possible
        
        for s_type in self.service.service_type: # agent needs to be qualified to perform the service
            if(s_type not in self.agent.qualifications):
                return False
        
        return True
    
    # TODO: check scoring function: very important & based on arbitrary scaling
    def similarity_score(self):
        sim = 0
        # score that reflects how similar a service and an agent are. A higher similarity score obviously means a more 'desirable pairing'
        if self.agent.age < (self.service.age - 40):
            sim -= 3
        elif self.agent.age > (self.service.age - 20):
            sim += 6
        
        if self.agent.region == self.service.region:
            sim += 10
        else:
            sim -= 10
        
        if self.agent.gender == self.service.gender:
            sim += 10
        else:
            sim -= 10
        
        num_common_lang = len(set(self.service.languages) & set(self.agent.languages))
        if num_common_lang == 0:
            sim -= 50
        elif num_common_lang > 1:
            sim += 5
        
        if self.agent.race == self.service.race:
            sim += 5
        else:
            sim -= 5
            
        dist = self.dist
        quant25, quant50, quant75 = get_quantiles()
        
        if dist <= quant25:
            sim += 30
        elif dist <= quant50:
            sim += 10
        elif dist <= quant75:
            sim -= 10
        else:
            sim -= 30
        
        min_score = -98 # =(-3-10-50-5-30)
        max_score = 66 # =(6+10+10+5+5+30)
        
        norm_sim = (sim - min_score) / (max_score - min_score) # min-max normalization
        return norm_sim # [0;1]
    
    @property
    def dist(self):
        return int(distance(self.agent.location, self.service.location).km)

    def to_string(self):
        return "Pair of " + self.service.name + " and " + self.agent.name
```

A pair represents a match of an agent to a service.

```python
# builds all possible legal pairs: 

def build_all_legal_pairs():
    pairs = list()
    for agent in all_agents:
        for service in all_services:
            pair = Pair(agent = agent, service = service)
            if pair.isLegal():
                pairs.append(pair)
    return pairs
```

## State

```python
class State:
    def __init__(self, pairs):
        self.pairs = pairs
        self.state_id = self.__key()
        self.services, self.agents = list(), list()
        for pair in self.pairs:
            self.services.append(pair.service)
            self.agents.append(pair.agent)
        
        self.assigned_hours = dict()
        for agent in all_agents:
            self.assigned_hours[agent] = 0
    
    def __key(self):
        key = []
        for pair in self.pairs:
            key.append(pair.pair_id,)
        return hash(tuple(key))
    
    def __eq__(self, other):
        if isinstance(other, State):
            return self.state_id == other.state_id
        return NotImplemented
    
    def __hash__(self):
        return self.state_id
    
    @property
    def total_dist(self):
        total_dist = 0
        for pair in self.pairs:
            total_dist += pair.dist
        return total_dist
    
    def isLegal(self):
        # all pairs must be legal
        for pair in self.pairs:
            if pair.isLegal() == False:
                return False
            
        # all services must be fulfilled
        if len(all_services) != len(self.services):
            return False
        for service in all_services:
            if service not in self.services:
                return False
        return True
    
    def similarity_score(self):
        sim = 0
        for pair in self.pairs:
            sim += pair.similarity_score()
        sim = sim / len(self.pairs)
        return sim
    
    def consistency_score(self, previous_state):
        sim = 0
        for pair in self.pairs:
            if pair in previous_state.pairs:
                sim += 1
        sim = sim / len(self.pairs)
        return sim
    
    @property
    def normal_hours(self):
        num_agents = len(self.agents)
        return self.agents[0].normal_hours * num_agents
    
    @property
    def total_hours(self):
        total_hours = 0
        for agent in self.agents:
            total_hours += agent.worked_hours
        return total_hours
    
    @property
    def active(self):
        for agent in self.agents:
            if agent.active == False:
                return False
        for service in self.services:
            if service.active == False:
                return False
        return True
    
    def assign(self, agent, num_hours):
        self.assigned_hours[agent] = num_hours
    
    def to_string(self):
        s=""
        for pair in self.pairs:
            s += pair.to_string()
            s += "; "
        return s[0:len(s)-1]
```

A state represents a combination of legal agent-service pairs where all services are fulfilled.

```python
def build_matches():
    # key: services
    # value: list of which eligible agents for this service
    
    reset_agents()
    all_legal_pairs = build_all_legal_pairs()
    matches = dict()
    
    for service in all_services:
        match = list()
        for pair in all_legal_pairs:
            if pair.service == service:
                match.append(pair.agent)
        matches[service] =  match
        
    return matches
```

```python
matches = build_matches()
```

Matches reflect a summary of all possible eligible agents for every given service.

```python
def build_state_space():
    reset_agents()
    matches = build_matches()
    n=0 # safety to avoid endless while loop

    pointers = dict()
    for service in matches.keys():
        pointers[service] = 0 
    
    state_space = list()
    done = False
    while(done == False and n < 100000):
        n+=1
        pairs = list()
        considered_agents, considered_time_slots = list(), list()
        #print()
        #print("------------------------------------------------")
        #print()
        #print("Iteration number {}".format(n))
        for service in matches.keys():
            #print("Now considering: {}".format(service.to_string()))
            service_pointer = pointers[service]
            #print("Pointer for this service: {}".format(service_pointer))
            considered_agent = matches[service][service_pointer]
            potential_pair = Pair(agent = considered_agent, service = service)
            #print("Considering potential pair {}".format(potential_pair.to_string()))
            
            if potential_pair.isLegal(): # success
                #print("This pair is legal!")
                pairs.append(potential_pair)
                #print("{} availabilties before are: {}".format(considered_agent.to_string(), considered_agent.current_availabilities))
                considered_agent.assign(service.time_slot)
                #print("{} availabilties are now: {}".format(considered_agent.to_string(), considered_agent.current_availabilities))
                considered_agents.append(considered_agent)
                considered_time_slots.append(service.time_slot)
                
                if(service == list(matches)[-1]): # last service was successfully allocated: total success
                    #print("The last service was successfully allocated!")
                    state = State(pairs)
                    
                    for agent in state.agents:
                        num_hours = (len(agent.availabilities) - len(agent.current_availabilities))/2
                        state.assign(agent, num_hours)
                    
                    state_space.append(state)
                    idx = list(matches.keys()).index(service)
                    pointers, done = update_pointers(idx, service, pointers, matches)
                        
                    reset_agents()
                    for agent in all_agents:
                        agent.reset_availabilities()
                    #print("Updated pointers are now: {}".format(pointers))
                    if done:
                        break

            else: # failure
                #print("This pair is NOT legal!")
                idx = list(matches.keys()).index(service)
                pointers, done = update_pointers(idx, service, pointers, matches)
                #print("Updated pointers are now: {}".format(pointers))
                rollback_slots(agents=considered_agents, time_slots=considered_time_slots)
                break
                
    return state_space

def update_pointers(idx, service, pointers, matches): # updates pointer of service
    if pointers[service] == (len(matches[service])-1): # pointer points to last eligible agent for service
        if idx == 0: # there are no more routes to consider
            return None, True
        else:
            pointers[service] = 0
            s = list(matches.keys())[idx-1]
            return update_pointers(idx-1, s, pointers, matches)
    
    else: # still eligible agents left for this service
        pointers[service] += 1
        return pointers, False

def rollback_slots(agents, time_slots):
    for idx in range(len(agents)):
        #print("{} cur_av before rollback: {}".format(agents[idx].to_string(), agents[idx].current_availabilities))
        agents[idx].current_availabilities.extend(time_slots[idx])
        agents[idx].current_availabilities.sort()
        #print("{} cur_av after rollback: {}".format(agents[idx].to_string(), agents[idx].current_availabilities))
```

## Action

An action represents going from one state to the next.

The action space equals the state space because every state is reachable independent of the state at the previous timestep.

## Reward

```python
def calc_reward(state, previous_state, end_of_month):
    WEIGHTS = [0.2, # weight similarity
               0.3, # weight consistency
               0.3, # weight financial impact
               0.2 # weight overtime/idle
              ]
    reward = 0
    
    #[4,4,1,1]
    #2,3,3,2
    
    #fin_score = (state.total_hours() - previous_state.total_hours()) / (previous_state.total_hours()) # percentage change of hours
    
    overtime, idle = 0, 0
    for agent in all_agents:
        #print("Overtime is", agent.overtime)
        #print("Idle is", agent.idle)
        if agent in state.agents:
            diff = state.assigned_hours[agent] - agent.normal_hours
            #print("Diff is", diff)
            if diff > 0: # net overtime for agent
                overtime += diff
            else: # net idle for agent
                idle += abs(diff)
        else:
            idle = agent.normal_hours
    
    num_agents = len(all_agents)
    
    max_overtime = num_agents*2
    min_overtime = 0
        
    max_idle = num_agents*2
    min_idle = 0
        
    overtime_score = (overtime - min_overtime) / (max_overtime - min_overtime)
    idle_score = (idle - min_idle) / (max_idle - min_idle)
    
    fin_score = (overtime_score + idle_score) / 2
    
    cons_score = state.consistency_score(previous_state) # [0;1]
    #print("State total_hours is {} and in previous state {}".format(state.total_hours(), previous_state.total_hours()))
    #print("This amounts to a fin_score of", fin_score)
    #print("The consistency score is", cons_score)
       
    sim_score = state.similarity_score() # [0;1]
    #print("The similarity is", sim_score)
    reward += sim_score * WEIGHTS[0]
    reward += cons_score * WEIGHTS[1]
    reward -= fin_score * WEIGHTS[2]
    
    if end_of_month:
        # extra penalties for idle/overtime
        overtime, idle = 0, 0
        #print()
        #print('Its the end of the month')
        #print()
        for agent in all_agents:
            diff = agent.overtime - agent.idle
            #print("Diff is", diff)
            if diff > 0: # net overtime for agent
                overtime += diff
            else: # net idle for agent
                idle += abs(diff)
        
        #print("Total overtime", overtime)
        #print("Total idle", idle)
        
        max_overtime = num_agents*60
        min_overtime = 0
        
        max_idle = num_agents*60
        min_idle = 0
        
        overtime_score = (overtime - min_overtime) / (max_overtime - min_overtime)
        idle_score = (idle - min_idle) / (max_idle - min_idle) 
        
        #print("Overtime score is", overtime_score)
        #print("Idle score is", idle_score)
        diff_score = (overtime_score + idle_score) / 2
        #print("Diff score is", diff_score)
        
        reward -= diff_score * WEIGHTS[3]
        
        # process end of month
        for agent in all_agents:
            agent.overtime = 0
            agent.idle = 0
    
    return reward
```

## Environment

```python
class Dynamic(gym.Space):
    """
    x where x in available actions {0,1,3,5,...,n-1}
    Example usage:
    self.action_space = spaces.Dynamic(max_space=2)
    """

    def __init__(self, max_space):
        self.n = max_space
        #initially all actions are available
        self.available_actions = list(range(0, max_space))

    def disable_actions(self, actions):
        self.available_actions = [action for action in self.available_actions if action not in actions]
        l = list(set(self.available_actions))
        l.sort()
        return l

    def enable_actions(self, actions):
        self.available_actions.extend(actions)
        l = list(set(self.available_actions))
        l.sort()
        return l

    def sample(self):
        return np.random.choice(self.available_actions)

    def contains(self, x):
        return x in self.available_actions
    
    @property
    def shape(self):
        return (0, len(self.available_actions))

    def __repr__(self):
        return "Dynamic(%d)" % self.n

    def __eq__(self, other):
        return self.n == other.n
```

```python
class Environment(Env):
    def __init__(self):
        self.action_space = Discrete(len(state_space))
        self.observation_space = Discrete(len(state_space))
        self.legal_action_space = Dynamic(max_space = len(state_space)) # custom space
        
        self.state = self.observation_space.sample()
        self.previous_state = None
        self.num_actions = 0
        self.timer = 61
    
    def step(self, action, change_prob=0.01):
        if action not in self.legal_action_space: # only legal actions should performed (-> no inactive agents/services)
            raise Exception
        
        self.previous_state = self.state
        self.state = action
        
        #print("State is", self.state)
        #print("Previous state is", self.previous_state)
        
        #print("Sample agent {} worked {} hours before".format(state_idx[self.state].agents[0].to_string(), state_idx[self.state].agents[0].worked_hours))
        
        # every agent in the current state works
        for agent in all_agents:
            num_hours = state_idx[self.state].assigned_hours[agent]
            agent.work(num_hours)
        
        #print("Agents have worked")
        #print("Sample agent {} worked {} hours after".format(state_idx[self.state].agents[0].to_string(), state_idx[self.state].agents[0].worked_hours))
        
        #print("In current state {} hours worked after".format(state_idx[self.state].total_hours()))
        
        # randomly en-/disable agents & services + update legal action space accordingly
        self.transition()
        
        end_of_month = (self.num_actions % 30 == 0)

        reward = calc_reward(state=state_idx[self.state], 
                             previous_state=state_idx[self.previous_state], 
                             end_of_month = end_of_month)
        
        if self.timer <= 0:
            done = True
        else:
            done = False
        
        self.num_actions += 1
        self.timer -= 1
        info = {}
        
        return self.state, reward, done, info
    
    def render(self):
        #TODO: insert visualization!
        pass
    
    def reset(self):
        self.legal_action_space = Dynamic(max_space = len(state_space))
        
        self.state = self.observation_space.sample()
        self.previous_state = None
        self.num_actions = 0
        self.timer = 61
        
        reset_agents()
        
        return self.state
    
    def transition(self, change_prob=0.001):
        change = False
        
        for agent in all_agents:
            if random() < change_prob: # change_prob probability of happening
                agent.active = not agent.active
                change = True
                
        for service in all_services:
            if random() < change_prob: # change_prob probability of happening
                service.active = not service.active
                change = True
        
        if change:
            #print("A change has occurred")
            self.changed()
    
    def changed(self):
        to_enable, to_disable = [], []
        for state in range(self.observation_space.n):
            if state_idx[state].active:
                if state not in self.legal_action_space:
                    # state is fully active but not in legal actions => enable
                    to_enable.append(state)
            else:
                if state in self.legal_action_space:
                    # state is NOT (fully) active but in legal actions => disable
                    to_disable.append(state)
        #print("We are enabling {} states and disabling {} states".format(len(to_enable), len(to_disable)))            
        self.legal_action_space.enable_actions(to_enable)      
        self.legal_action_space.disable_actions(to_disable)         
```

# Reinforcement Learning Algorithm

## Q Learning

```python
def q_train(env, num_episodes, gamma = 0.9, init_lr = 0.6, init_epsilon = 0.3, end_epsilon = 0.05): # train Q table
    t1 = time()
    q_table = np.zeros((env.observation_space.n, env.action_space.n)) # row = state, col = action
    
    history = {
        'rewards': np.zeros(num_episodes)
              }
    
    epsilon = init_epsilon # initialize epsilon
    lr = init_lr
    
    for episode in range(num_episodes):
        state = env.reset()
        score = 0
        done = False
        epsilon = update_epsilon(init_epsilon, end_epsilon, episode, num_episodes) # epsilon decay
        lr = update_lr(init_lr, episode)
        
        while done == False:
            #env.render()
            if random() < epsilon: # epsilon probability of exploration
                action = env.legal_action_space.sample()
            else: # (1 - epsilon) probability of exploitation
                action = best_legal_action(state, q_table)

            #print("Take action {}".format(action))
            next_state, reward, done, info = env.step(action)
            score += reward
            
            history['rewards'][episode] = score
            
            old_value = q_table[state][action]
            next_max = np.max(q_table[next_state])
            
            td_target = reward + gamma * next_max
            td = td_target - old_value
            q_table[state][action] += lr * td
   
            # done is True if episode terminated   
            if done:
                break
            
            state = next_state
    
    
    print(f'training finished: {time() - t1} seconds')
    return q_table, history
    
def update_epsilon(init_epsilon, end_epsilon, episode, N):
    # decayed epsilon greedy
    return init_epsilon
#     r = max( ((N - episode) / N) , 0)
#     new_epsilon = (init_epsilon - end_epsilon) * r + end_epsilon
#     return new_epsilon

def update_lr(init_lr, episode, k=0.1):
    return init_lr
    #return init_lr * math.exp(-k * episode)

def best_legal_action(state, q_table):
    legal_q_table = np.take(q_table, env.legal_action_space.available_actions, axis=1) # filter out illegal columns (actions)
    return np.argmax(legal_q_table[state])
```

```python
def q_perform(env, q_table, num_episodes): # take optimal actions according to Q table
    history = {
        'rewards': np.zeros(num_episodes)
              }
    for episode in range(num_episodes):
        state = env.reset()
        score = 0
        done = False
        
        
        while done == False:
            #env.render()
            legal_q_table = np.take(q_table, env.legal_action_space.available_actions, axis=1) # filter out illegal columns (actions)
            action = np.argmax(legal_q_table[state])

            #print("Take action {}".format(action))
            next_state, reward, done, info = env.step(action)
            score += reward
            history['rewards'][episode] = score
   
            # done is True if episode terminated   
            if done:
                break
            
            state = next_state
    
    return history
```

Learning rate decay: https://www.jmlr.org/papers/volume5/evendar03a/evendar03a.pdf

## Deep Q Learning

```python
env = Environment()
```

```python
def build_nn(states, actions):
    model = Sequential()
    model.add(Dense(32, activation='relu', input_shape=states))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(actions, activation='linear')) # Output: Q value estimate
    return model
```

```python
model = build_nn((env.observation_space.n,), env.action_space.n)
model.summary()
```

```python
def build_dqn(model, actions):
    policy = EpsGreedyQPolicy(eps=0.1) # Epsilon-Greedy policy for exploration vs exploitation trade-off
    memory = SequentialMemory(limit=5000, window_length=actions)
    dqn = DQNAgent(model=model, memory=memory, policy=policy, nb_actions=actions, nb_steps_warmup=actions+1, target_model_update=1e-2)
    return dqn
```

```python
dqn = build_dqn(model, env.legal_action_space.n)
```

```python
dqn.compile(Adam(lr=1e-3), metrics=['mae'])
dqn.fit(env, nb_steps=30000, visualize=False, verbose=1)
```

```python
dqn_scores = dqn.test(env, nb_episodes=20, visualize=False)
```
